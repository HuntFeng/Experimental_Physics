# Introduction to matplotlib.pyplot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# prepare a sequence of x
x = np.arange(-1,1,0.5)
print(x)
```

```python
x = np.linspace(-1,1,5)
print(x)
```

```python
# prepare a sequence of y
# say y = x^2-1
y = x**2 - 1
print(y)
```

## `figure(num=None, figsize=None)`
* num(optional): number of the figure
* figsize(optional): figure size

## `plot(x, y, '', linewidth=None, markersize=None)`
* '' is for changing the line style
* linewidth(optional): width of the line
* markersize(optional): size of the markers

## `xlabel(xlabel, fontsize=None)` and `ylabel`
## `title(title,fontsize=None)`

```python
plt.figure(figsize=(6,4))
plt.plot(x,y,'ro--',linewidth=3,markersize=10)
plt.xlabel("x")
plt.ylabel("f(x)", fontsize=20)
plt.title("A quadratic function", fontsize=30)
plt.show()
```

## `errorbar(x, y, yerr=None, xerr=None, fmt='', ecolor=None, elinewidth=None, capsize=None, capthick=None)`

* fmt is for the line style
* the rest is for the errorbar style

```python
x = np.linspace(1,5,5) # x = np.array([1,2,3,4,5])
y = [2.1,3.9,6.5,7.3,10]
y_error = [0.5,0.5,0.5,0.5,0.5]
```

```python
plt.errorbar(x,y,yerr=y_error,
             fmt='o',ecolor='r',capsize=5,
             label='measurements')
plt.plot(x,2*x,label='best-fit line')
plt.legend(fontsize=15)
plt.title("A title", fontsize=30)
plt.xlabel("x",fontsize=15)
plt.ylabel("y",fontsize=15)
plt.savefig("Some figure.png")
plt.show()
```
